```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Display settings
pd.set_option('display.max_columns', 100)
pd.set_option('display.max_rows', 100)
pd.set_option('display.float_format', lambda x: '%.3f' % x)

# Plot settings
plt.rcParams['figure.figsize'] = (12, 6)
plt.rcParams['figure.dpi'] = 100
sns.set_style('whitegrid')

print("All imports successful")
```

```python
df = pd.read_csv('../data/raw/application_train.csv')

print(f"Rows    : {df.shape[0]:,}")
print(f"Columns : {df.shape[1]}")
print(f"\nFirst look at the data:")
df.head()
```

```python
target_counts = df['TARGET'].value_counts()
target_pct = df['TARGET'].value_counts(normalize=True) * 100

print("Target Variable Distribution")
print("="*40)
print(f"Repaid    (0) : {target_counts[0]:,}  ({target_pct[0]:.2f}%)")
print(f"Defaulted (1) : {target_counts[1]:,}  ({target_pct[1]:.2f}%)")
print(f"\nClass Imbalance Ratio : {target_counts[0]/target_counts[1]:.1f}:1")
```

```python
fig, axes = plt.subplots(1 , 2, figsize = (14 , 5))

# Bar chart
axes[0].bar(['Repaid (0)' , 'Defaulted (1)'], target_counts, 
            color=['steelblue' , 'tomato'] , edgecolor = 'black' , width = 0.5)
axes[0].set_title('Target Variable Count' , fontsize = 14 , fontweight = 'bold')
axes[0].set_ylabel('Number of Applicants')
for i, v in enumerate(target_counts):
    axes[0].text(i , v + 1000 , f'{v:,}', ha = 'center' , fontweight = 'bold')

# Pie chart
axes[1].pie(target_counts , labels=['Repaid (0)' , 'Defaulted (1)'] ,
            colors=['steelblue' , 'tomato'], autopct='%1.2f%%' ,
            startangle = 90 , explode=(0 , 0.05))
axes[1].set_title('Target Variable Percentage' , fontsize = 14 , fontweight ='bold')

plt.tight_layout()
plt.savefig('../docs/target_distribution.png' , bbox_inches='tight')
plt.show()

print("\nPlot saved to docs/target_distribution.png")
```

```python
numerical_cols = df.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_cols = df.select_dtypes(include=['object']).columns.tolist()

print("Data Types Overview")
print("= =" * 40)
print(f"Total Columns      : {df.shape[1]}")
print(f"Numerical Columns  : {len(numerical_cols)}")
print(f"Categorical Columns: {len(categorical_cols)}")

print("\nCategorical Columns:")
print("-"*40)
for col in categorical_cols:
    print(f"  {col:40s} — {df[col].nunique()} unique values")
```

```python
# Calculate missing values
missing = pd.DataFrame({
    'Missing Count': df.isnull().sum(),
    'Missing Percent': (df.isnull().sum() / len(df)) * 100
})

# Filter only columns that have missing values and sort
missing = missing[missing['Missing Count'] > 0].sort_values('Missing Percent', ascending=False)

print(f"Total columns with missing values: {len(missing)}")
print(f"Total columns with >30% missing  : {len(missing[missing['Missing Percent'] > 30])}")
print(f"Total columns with >50% missing  : {len(missing[missing['Missing Percent'] > 50])}")
print("\nTop 30 columns by missing percentage:")
print("="*55)
print(f"{'Column':<45} {'Missing %':>8}")
print("-"*55)
for col, row in missing.head(30).iterrows():
    print(f"{col:<45} {row['Missing Percent']:>7.2f}%")

# Plot top 30 missing columns
plt.figure(figsize=(14, 10))
top_missing = missing.head(30)
bars = plt.barh(top_missing.index, top_missing['Missing Percent'], 
                color='tomato', edgecolor='black')
plt.axvline(x=30, color='black', linestyle='--', linewidth=1.5, label='30% threshold')
plt.axvline(x=50, color='aqua', linestyle='--', linewidth=1.5, label='50% threshold')
plt.xlabel('Missing Percentage')
plt.title('Top 30 Columns by Missing Value Percentage', fontsize=14, fontweight='bold')
plt.legend()
plt.gca().invert_yaxis()
plt.tight_layout()
plt.savefig('../docs/missing_values.png', bbox_inches='tight')
plt.show()

print("\nPlot saved to docs/missing_values.png")
```

```python
cat_features = [
    'NAME_CONTRACT_TYPE',
    'CODE_GENDER',
    'FLAG_OWN_CAR',
    'FLAG_OWN_REALTY',
    'NAME_INCOME_TYPE',
    'NAME_EDUCATION_TYPE',
    'NAME_FAMILY_STATUS',
    'NAME_HOUSING_TYPE'
]

# Plot default rate for each categorical feature
fig, axes = plt.subplots(4, 2, figsize=(18, 24))
axes = axes.flatten()

for i, col in enumerate(cat_features):
    default_rate = df.groupby(col)['TARGET'].mean().sort_values(ascending=False) * 100
    
    bars = axes[i].bar(default_rate.index, default_rate.values,
                       color='steelblue', edgecolor='black')
    axes[i].set_title(f'Default Rate by {col}', fontsize=12, fontweight='bold')
    axes[i].set_ylabel('Default Rate (%)')
    axes[i].tick_params(axis='x', rotation=30)
    
    # Add value labels on bars
    for bar, val in zip(bars, default_rate.values):
        axes[i].text(bar.get_x() + bar.get_width()/2, 
                     bar.get_height() + 0.1,
                     f'{val:.1f}%', ha='center', fontsize=9, fontweight='bold')

plt.suptitle('Default Rate by Categorical Features', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('../docs/default_rate_categorical.png', bbox_inches='tight')
plt.show()

print("Plot saved to docs/default_rate_categorical.png")
```

```python
numerical_features = {
    'AGE'              : 'DAYS_BIRTH',
    'YEARS_EMPLOYED'   : 'DAYS_EMPLOYED',
    'INCOME'           : 'AMT_INCOME_TOTAL',
    'CREDIT_AMOUNT'    : 'AMT_CREDIT',
    'ANNUITY'          : 'AMT_ANNUITY'
}

# DAYS_BIRTH is negative (days before application) — convert to years
df['AGE_YEARS'] = (-df['DAYS_BIRTH'] / 365).astype(int)

# DAYS_EMPLOYED is negative for employed people — some have 365243 which means unemployed
df['DAYS_EMPLOYED_CLEAN'] = df['DAYS_EMPLOYED'].replace(365243, np.nan)
df['YEARS_EMPLOYED_CLEAN'] = (-df['DAYS_EMPLOYED_CLEAN'] / 365)

fig, axes = plt.subplots(5, 1, figsize=(14, 30))

# --- Age ---
df['AGE_BIN'] = pd.cut(df['AGE_YEARS'], bins=[18, 25, 35, 45, 55, 65, 100],
                        labels=['18-25', '26-35', '36-45', '46-55', '56-65', '65+'])
age_default = df.groupby('AGE_BIN', observed=True)['TARGET'].mean() * 100
axes[0].bar(age_default.index.astype(str), age_default.values, color='steelblue', edgecolor='black')
axes[0].set_title('Default Rate by Age Group', fontsize=13, fontweight='bold')
axes[0].set_ylabel('Default Rate (%)')
for i, v in enumerate(age_default.values):
    axes[0].text(i, v + 0.1, f'{v:.1f}%', ha='center', fontweight='bold')

# --- Years Employed ---
df['EMP_BIN'] = pd.cut(df['YEARS_EMPLOYED_CLEAN'], bins=[0, 1, 3, 5, 10, 20, 100],
                        labels=['<1yr', '1-3yr', '3-5yr', '5-10yr', '10-20yr', '20+yr'])
emp_default = df.groupby('EMP_BIN', observed=True)['TARGET'].mean() * 100
axes[1].bar(emp_default.index.astype(str), emp_default.values, color='steelblue', edgecolor='black')
axes[1].set_title('Default Rate by Years Employed', fontsize=13, fontweight='bold')
axes[1].set_ylabel('Default Rate (%)')
for i, v in enumerate(emp_default.values):
    axes[1].text(i, v + 0.1, f'{v:.1f}%', ha='center', fontweight='bold')

# --- Income ---
df['INCOME_BIN'] = pd.qcut(df['AMT_INCOME_TOTAL'], q=5,
                            labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'],
                            duplicates='drop')
income_default = df.groupby('INCOME_BIN', observed=True)['TARGET'].mean() * 100
axes[2].bar(income_default.index.astype(str), income_default.values, color='steelblue', edgecolor='black')
axes[2].set_title('Default Rate by Income Bracket', fontsize=13, fontweight='bold')
axes[2].set_ylabel('Default Rate (%)')
for i, v in enumerate(income_default.values):
    axes[2].text(i, v + 0.1, f'{v:.1f}%', ha='center', fontweight='bold')

# --- Credit Amount ---
df['CREDIT_BIN'] = pd.qcut(df['AMT_CREDIT'], q=5,
                             labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'],
                             duplicates='drop')
credit_default = df.groupby('CREDIT_BIN', observed=True)['TARGET'].mean() * 100
axes[3].bar(credit_default.index.astype(str), credit_default.values, color='steelblue', edgecolor='black')
axes[3].set_title('Default Rate by Credit Amount', fontsize=13, fontweight='bold')
axes[3].set_ylabel('Default Rate (%)')
for i, v in enumerate(credit_default.values):
    axes[3].text(i, v + 0.1, f'{v:.1f}%', ha='center', fontweight='bold')

# --- Annuity ---
df['ANNUITY_BIN'] = pd.qcut(df['AMT_ANNUITY'], q=5,
                              labels=['Very Low', 'Low', 'Medium', 'High', 'Very High'],
                              duplicates='drop')
annuity_default = df.groupby('ANNUITY_BIN', observed=True)['TARGET'].mean() * 100
axes[4].bar(annuity_default.index.astype(str), annuity_default.values, color='steelblue', edgecolor='black')
axes[4].set_title('Default Rate by Annuity Amount', fontsize=13, fontweight='bold')
axes[4].set_ylabel('Default Rate (%)')
for i, v in enumerate(annuity_default.values):
    axes[4].text(i, v + 0.1, f'{v:.1f}%', ha='center', fontweight='bold')

plt.suptitle('Default Rate by Numerical Features (Binned)', fontsize=16, fontweight='bold', y=1.01)
plt.tight_layout()
plt.savefig('../docs/default_rate_numerical.png', bbox_inches='tight')
plt.show()

print("Plot saved to docs/default_rate_numerical.png")
```

```python
key_numerical = [
    'TARGET',
    'AGE_YEARS',
    'AMT_INCOME_TOTAL',
    'AMT_CREDIT',
    'AMT_ANNUITY',
    'AMT_GOODS_PRICE',
    'REGION_POPULATION_RELATIVE',
    'DAYS_REGISTRATION',
    'OWN_CAR_AGE',
    'CNT_FAM_MEMBERS',
    'EXT_SOURCE_1',
    'EXT_SOURCE_2',
    'EXT_SOURCE_3',
    'DAYS_ID_PUBLISH',
    'DAYS_LAST_PHONE_CHANGE'
]

# Compute correlation matrix
corr_matrix = df[key_numerical].corr()

# Plot heatmap
plt.figure(figsize=(16, 12))
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))  # hide upper triangle
sns.heatmap(
    corr_matrix,
    mask=mask,
    annot=True,
    fmt='.2f',
    cmap='RdYlGn',
    center=0,
    linewidths=0.5,
    annot_kws={'size': 9},
    vmin=-1,
    vmax=1
)
plt.title('Correlation Heatmap — Key Numerical Features', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.savefig('../docs/correlation_heatmap.png', bbox_inches='tight')
plt.show()

print("Plot saved to docs/correlation_heatmap.png")
```

```python
missing_cols = missing[
    (missing['Missing Percent'] >= 30) & 
    (missing['Missing Percent'] <= 70)
].index.tolist()

# Take a sample of 5000 rows for visualization
sample_df = df[missing_cols].sample(n=5000, random_state=42)

# Plot missing value heatmap
plt.figure(figsize=(18, 8))
sns.heatmap(
    sample_df.isnull(),
    cbar=False,
    cmap='viridis',
    yticklabels=False
)
plt.title('Missing Value Patterns (Sample of 5000 rows)', fontsize=14, fontweight='bold')
plt.xticks(rotation=90, fontsize=7)
plt.tight_layout()
plt.savefig('../docs/missing_value_heatmap.png', bbox_inches='tight')
plt.show()

print("Plot saved to docs/missing_value_heatmap.png")

# Check if missingness is correlated with default
print("\nIs missingness correlated with default?")
print("="*50)
for col in ['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3', 'OWN_CAR_AGE', 'OCCUPATION_TYPE']:
    missing_mask = df[col].isnull()
    default_rate_missing  = df[missing_mask]['TARGET'].mean() * 100
    default_rate_present  = df[~missing_mask]['TARGET'].mean() * 100
    print(f"\n{col}")
    print(f"  Default rate when MISSING : {default_rate_missing:.2f}%")
    print(f"  Default rate when PRESENT : {default_rate_present:.2f}%")
```

## Key Insights from Default Rate Analysis

- Younger applicants **(18–25)** default more than older ones
- **Lower income brackets** default more
- **Less employment history** = higher default risk
- `EXT_SOURCE` features are **strong predictors** (external credit scores)
- Missing `EXT_SOURCE_3` = higher default risk (**9.31% vs 7.77%**)
